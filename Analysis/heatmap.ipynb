```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
```

```python
table = pd.read_excel('../data/for_heatmap.xls')
```

```python
vertical = table.pop('Unnamed: 0')
horisontal = list(table.head())
table = table.to_numpy()
```

```python
plt.rcParams['axes.titlesize'] = 30
fig = plt.figure(1, figsize=(30, 12))
gs = gridspec.GridSpec(1, 1)
ax0 = fig.add_subplot(gs[0, :])
number = sns.heatmap(table, xticklabels=horisontal, yticklabels=vertical, cmap='Blues_r', square= True)
bottom, top = number.get_ylim()
number.set_ylim(bottom + 0.5, top - 0.5)
plt.xlabel("Genes", fontsize=18)
plt.ylabel("Spicies", fontsize=18)
ax0.set_title('Pseudogenes heatmap')
#plt.savefig('/heatmap_blues.png', dpi=300)
plt.show()
```
